```python
#importing BeautifulSoup for HTML scraping
#importing ScraperAPIClient for implementing proxy server 

import requests
from bs4 import BeautifulSoup
from scraper_api import ScraperAPIClient
import pandas as pd


#reference used - https://towardsdatascience.com/scraping-multiple-amazon-stores-with-python-5eab811453a8
```

```python
#scraping primary product data

tracker_log = pd.DataFrame()
dtracker = pd.DataFrame()
ctracker = pd.DataFrame()

# imports a csv file with the url's to scrape
prod = pd.read_csv('C:\\Users\\azizk\\product_urls.csv', sep=',')
prod_URLS = prod.url

client = ScraperAPIClient('023c31e0153eb7532f2822c4fb1079a2')

###### for loop - keeps cycling through url's
for x, url in enumerate(prod_URLS):
    page = client.get(url, headers = {"X-MyHeader": "123"}).text
    soup = BeautifulSoup(page,'html.parser')
    
    title = soup.find(id='productTitle').get_text().strip()
    
    seller = soup.find(id='bylineInfo').get_text().strip()
    
    price = soup.find(id='priceblock_ourprice').get_text().replace('\xa0',' ').strip()
    
    try:
        description = soup.find(id='productDescription').get_text().strip()
        
    except:
        description = 'No description available'
    
    rating = float(soup.select('.a-star-4-5')[0].get_text().split(' ')[0].replace(",", "."))
    reviews = int(soup.select('#acrCustomerReviewText')[0].get_text().split(' ')[0].replace(",", ""))
    stock = soup.find(id='availability').get_text().strip().replace('\n',' ')
    
    log = pd.DataFrame({'url': url,
                        'title': title,
                        'price': price,
                        'stock': stock,
                        'description': description,
                        'rating': rating,
                        'reviews': reviews
                        'seller': seller
                        }, index = [x])
    dlog = pd.DataFrame({
                         'description': description,
                        })
    clog = pd.DataFrame({'category': categories,
                        })

    tracker_log = tracker_log.append(log)
    dtracker = dtracker.append(dlog)
    ctracker = ctracker.append(clog)
    print('appended '+'\n' + title + '\n\n')
```

```python
#scraping product features

tracker = pd.DataFrame()

# imports a csv file with the url's to scrape
product = pd.read_csv('C:\\Users\\azizk\\product_urls.csv', sep=',')
product_URLS = product.url

client = ScraperAPIClient('023c31e0153eb7532f2822c4fb1079a2')


###### function - keeps cycling through url's
for x, url in enumerate(product_URLS):
    page = client.get(url, headers = {"X-MyHeader": "123"}).text
    soup = BeautifulSoup(page,'html.parser')
    
    title = soup.find(id='productTitle').get_text().strip()
    
    features = []
    for li in soup.select("#feature-bullets ul.a-unordered-list")[0].findAll('li'):
        features.append(li.get_text().strip())
    
      
    log = pd.DataFrame({'title': title,
                        'features': features,
                       })
    
    tracker = tracker.append(log)
    print('appended '+'\n' + title + '\n\n')
```

```python
#scraping product categories

ptracker = pd.DataFrame()

# imports a csv file with the url's to scrape
product = pd.read_csv('C:\\Users\\azizk\\product_urls.csv', sep=',')
product_URLS = product.url

client = ScraperAPIClient('023c31e0153eb7532f2822c4fb1079a2')


###### function - keeps cycling through url's
for x, url in enumerate(product_URLS):
    page = client.get(url, headers = {"X-MyHeader": "123"}).text
    soup = BeautifulSoup(page,'html.parser')
    title = soup.find(id='productTitle').get_text().strip()
 
    categories = []
    for li in soup.select("#wayfinding-breadcrumbs_container ul.a-unordered-list")[0].findAll("li"):
        categories.append(li.get_text().strip())
        if '\u203a' in categories:
            categories.remove('\u203a')
    
    log = pd.DataFrame({'title': title,
                        'category': categories,
                        })
    
    ptracker = ptracker.append(log)
    print('appended '+'\n' + title + '\n\n')
```
